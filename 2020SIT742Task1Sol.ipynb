```python
lower = []
for item in df_text['job_description']:
    lower.append(item.lower())           # lowercase description

listToStr = ' '.join([str(elem) for elem in lower]) 
type(listToStr) 
listToStr   

tokenizer = RegexpTokenizer(r"\w+(?:[-']\w+)?")
tokens = tokenizer.tokenize(listToStr)
tokens

```

Secondly, 

<div class="alert alert-block alert-info">

```python
listToStr1 = ' '.join([str(elem) for elem in tokens]) 
type(listToStr1) 
listToStr1  

stop_words = set(stopwords.words('english')) 
token = word_tokenize(listToStr1)
result = [i for i in token if not i in stop_words]
result

cts = nltk.FreqDist(result)
df_cts = pd.DataFrame.from_dict(cts, orient='index')
df_cts.columns = ['Frequency']
df_cts.columns.name = 'words'
freq6000 = df_cts[(df_cts['Frequency'] >= 6000)]
freq6000.sort_values(by = ['Frequency'])
cts
```
